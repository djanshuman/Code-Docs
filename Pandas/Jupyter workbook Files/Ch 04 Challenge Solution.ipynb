Which country has won the most Men's Gold Medals in Singles Badminton over the years , Sort results alphabetically by Player name ?

```python
x1=oo[(oo.Sport=='Badminton') & (oo.Medal == 'Gold') & (oo.Gender == 'Men') &(oo.Event== 'singles')]
x1.sort_values(by='Athlete')
```

Which three countries have won the most Medals in recent years (1984-2008) ?

```python
rec=oo[(oo.Edition>=1984) & (oo.Edition <=2008)]
rec.NOC.value_counts().head(3)
```

Display the male Gold medal winners for 100m Track & Field sprint event over the years.List the results starting with most recent. Show the Olympic City,Edition,Athlete and Country they represent ?

```python
x1=oo[(oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Event == '100m')]
x1.sort_values(by='Edition',ascending=False)[['City','Edition','Athlete','NOC']]
```
