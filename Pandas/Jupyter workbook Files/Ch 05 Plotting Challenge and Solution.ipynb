Plot the number of medals achieved by the chinese team (men and women) in Beijing 2008 using:
. Matplotlib
. Seaborn

```python
x1=oo[(oo.NOC == 'CHN') & (oo.Edition == 2008)]
x1.Gender.value_counts().plot(kind='Bar');
```

```python
import seaborn as sns
```

```python
sns.countplot(x='Gender',data=x1);
```

How can you use Colormaps to give data more meaning?

```python
sns.countplot(x='Gender',data=x1,palette='bwr');
```

Plot the number of Gold, Silver and Bronze medals for each Gender?

```python
sns.countplot(x='Medal',data=x1,hue='Gender',palette='pastel');
```

How can you give the data more meaning? Is their anything else you can change to make it more intuitive.

```python
sns.countplot(x='Medal',data=x1,hue='Gender',palette='bwr',order=['Gold','Silver','Bronze'],);
```
