Using the olympic dataset, present a summary of the total medals won by participating countries in 2008 Olympics

```python
g=oo[oo.Edition==2008]
g1=g.groupby(['NOC','Medal']).size().unstack('Medal',fill_value=0)
g1.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
```

```python
sns.heatmap(g1)
```

```python
g1=g1.transpose()
g1
```

Creating your own Colormaps

```python
g = oo.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0)
g.sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']].head().plot(kind='bar')
```
